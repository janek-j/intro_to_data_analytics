## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
plt.scatter(houses_dataframe['sqft_living'], houses_dataframe['price'])
plt.xlabel('Square Feet Living Area')
plt.ylabel('Price')
plt.title('House Price <-> Square Feet Living Area')
plt.show()
```

```python
"""
Scatter plot illustrates the relationship between the square footage of the living area and the price of houses.
From the plot, we can observe a positive correlation: as the square footage of the living area increases, 
the price of the house tends to increase as well.
This suggests that larger homes generally command higher prices in the market.
"""
```

```python
#b)
plt.scatter(houses_dataframe['condition'], houses_dataframe['price'])
plt.xlabel('Condition')
plt.ylabel('Price')
plt.title('House Price <-> Condition')
plt.show()
```

```python
"""
Scatter plot illustrates the relationship between the condition of the house and its price.
From the plot, we can observe that houses in better condition (higher condition values) tend to have higher prices.
However, the relationship is not strictly linear, as there are some houses in lower conditions that also have relatively high prices.
Overall, the trend suggests that better-conditioned houses generally command higher prices in the market.
"""
```

```python
#c)
good_grade = houses_dataframe[houses_dataframe['grade'] >= 8]
plt.figure()
plt.scatter(good_grade['long'], good_grade['lat'], c=good_grade['price'], cmap='viridis', alpha=0.6)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Location of Good Grade Apartments (grade >= 8)')
plt.colorbar(label='Price')
plt.show()

plt.figure()
plt.scatter(houses_dataframe['yr_built'], houses_dataframe['price'])
plt.xlabel('Year Built')
plt.ylabel('Price')
plt.title('House Price <-> Year Built')
print("There is no correlation between year built and price :O")

plt.figure()
renovated = houses_dataframe[houses_dataframe['yr_renovated'] >= 1960]
plt.scatter(renovated['yr_renovated'], renovated['price'])
plt.xlabel('Year Renovated')
plt.ylabel('Price')
plt.title('House Price <-> Year Renovated (since 1960)')
plt.show()


```

```python
"""
Here we have 3 plots. The first plot shows the geographical distribution of houses with a grade of 8 or higher,
with color coding to represent the price of each house. It shows that in the middle are the most expensive houses.

The second plot shows the relationship between the year a house was built and its price, indicating no clear correlation.

The third plot shows the relationship between the year a house was renovated (since 1960) and its price, 
suggesting that renovations can impact house prices.
"""
```

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)

plt.figure(figsize=(10,6))
plt.boxplot([houses_dataframe[houses_dataframe['grade'] == g]['price'] for g in sorted(houses_dataframe['grade'].unique())],
            tick_labels=sorted(houses_dataframe['grade'].unique()))
plt.xlabel('Grade')
plt.ylabel('Price')
plt.title("Boxplot: Apartment's Grade vs Price")
plt.show()

#Boxplot shows that higher grade apartments tend to have higher prices, with some outliers

#The 3a plot shows that higher apartment grades are clearly associated with higher prices,
# while the 2b plot indicates that house condition has little effect on price.
```

```python
#b)

from statsmodels.graphics.mosaicplot import mosaic

price_bins = pd.cut(houses_dataframe['price'], bins=3, labels=['Low', 'Medium', 'High'])
houses_dataframe['price_category'] = price_bins

plt.figure(figsize=(10,7))
plt.rcParams.update({'font.size': 8}) 
mosaic(houses_dataframe, ['condition', 'price_category'], axes_label=True, properties=lambda key: {'color': 'lightblue'})
plt.title('Mosaic Plot: Condition vs Price Category')
plt.tight_layout()
plt.show()

#Mosaic plot shows that houses in better condition tend to fall into higher price categories.
```

```python
"""
This mosaic plot illustrates the relationship between house condition and price category.
Houses in better condition (higher condition values) tend to fall into higher price categories (Medium and High),
while those in poorer condition are more likely to be in the Low price category. 
"""
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)

selected = houses_dataframe.loc[
    (houses_dataframe['bedrooms'] == 3) &
    (houses_dataframe['bathrooms'] == 2) &
    (houses_dataframe['floors'] == 2)
]
display(selected.describe())
```

```python
"""
Display the statistical summary of houses with 3 bedrooms, 2 bathrooms, and 2 floors.
"""
```

```python
#b)

plt.figure(figsize=(8,6))
plt.scatter(houses_dataframe['sqft_living'], houses_dataframe['price'], alpha=0.3, label='All houses')
plt.scatter(selected['sqft_living'], selected['price'], color='red', label='Selected houses')
plt.xlabel('Square Feet Living Area')
plt.ylabel('Price')
plt.title('Selected Houses Highlighted')
plt.legend()
plt.show()

#There is a lot of houses between (0-5000) sqft living area, but selected houses are mostly between 1000-3000 sqft.
```

```python
"""
Here we can see that the more square feet living are is, the more expensive the house is.
Our specific houses ranges mostly between 1000-3000 sqft.
"""
```

```python
#c)

selected = selected.copy()
selected['price_per_sqft'] = selected['price'] / selected['sqft_living']
best = selected.sort_values('price_per_sqft')
display(best)

#The best house for our demands is house with id "7452500565"
```
